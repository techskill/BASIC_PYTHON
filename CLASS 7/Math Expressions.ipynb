## Math expressions: Eliminating ambiguity

```python
#example : 01
result = 10 + 5 * 2
print(result) # 20 or 30
```

```python
#example : 02
result = (10 + 5) * 2
print(result)
```

```python
#example : 03
price = 100
discount = 20
final = price - discount / 100 * price
print(final)
```

```python
#example : 04
final = price - (discount / 100 * price)
```
