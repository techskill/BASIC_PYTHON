# Condition Statements

```python
age = 18

if age >= 18:
    print("Adult")
```

```python
age = 16

if age >= 18:
    print("Adult")
else:
    print("Minor")
```

```python
marks = 75

if marks >= 80:
    print("Grade A")
elif marks >= 60:
    print("Grade B")
else:
    print("Grade C")
```

```python
temperature = 35

if temperature >= 40:
    print("Very Hot")
elif temperature >= 25:
    print("Warm")
else:
    print("Cold")
```
